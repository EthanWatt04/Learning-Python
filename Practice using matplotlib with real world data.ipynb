```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('us-counties.csv')

df.head(20)
```

```python
df_Washington = df[df['state'] == 'Washington'].copy()
```

```python
df_Washington['date'] = pd.to_datetime(df_Washington['date'])
```

```python
Washington_sum = df_Washington.groupby('date')['deaths'].sum()
```

```python
plt.plot(Washington_sum.index, Washington_sum.values)

plt.title("Washington")

plt.show()
```

```python
plt.figure(figsize=(10,5))

unique_states = df['state'].unique()

for state in unique_states:
    df_state = df[df['state'] == state].copy()
    df_state['date'] = pd.to_datetime(df_state['date'])
    state_sum = df_state.groupby('date')['deaths'].sum()
    
    plt.plot(state_sum.index, state_sum.values, label = state)



plt.show()
```
