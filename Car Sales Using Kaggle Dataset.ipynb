```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('car_sales.csv')

df.head()
```

```python
df_2 = df.groupby('Manufacturer')['Sales_in_thousands'].sum().reset_index()
df_3 = df_2[0:10]
df_3
```

```python
plt.figure(figsize=(25, 15))

plt.bar(df_3.Manufacturer, df_3.Sales_in_thousands, width = 0.80)

plt.xlabel('Car Manufacturer')
plt.ylabel('Sales')
plt.title('Total Number Of Sales Per Manufacturer')

plt.show()
```

```python
df['Manufacturer'].unique()
```

```python
df_Acura = df.loc[df['Manufacturer'] == 'Acura']
```

```python
df_Acura
```

```python
plt.figure(figsize=(25, 15))

plt.bar(df_Acura.Model, df_Acura.Sales_in_thousands, color='Black')

plt.xlabel('Model')
plt.ylabel('Number of Sales In Thousands')
plt.title('Number Of Sales Per Model')
plt.suptitle('Acura')


plt.show()
```

```python
plt.figure(figsize=(25, 15))

plt.plot(df_Acura.Engine_size, df_Acura.Horsepower, 'o--k')

plt.xlabel('Engine Size')
plt.ylabel('Horsepower')
plt.title('Engine Size vs Horsepower')
plt.suptitle('Acura')

plt.show()
```

```python
plt.figure(figsize=(25, 15))

plt.plot(df_Acura.Engine_size, df_Acura.Fuel_efficiency, 'o--k')

plt.xlabel('Engine Size')
plt.ylabel('Fuel Efficiency')
plt.title('Fuel Efficiency Per Engine Size')
plt.suptitle('Acura')

plt.show()
```

```python
plt.figure(figsize=(25, 15))

df_Volkswagen = df.loc[df['Manufacturer'] == 'Volkswagen']

plt.bar(df_Volkswagen.Model, df_Volkswagen.Sales_in_thousands, color='#1f2f57')

plt.xlabel('Model')
plt.ylabel('Number of Sales In Thousands')
plt.title('Number Of Sales Per Model')
plt.suptitle('Volkswagen')


plt.show()
```

```python
plt.figure(figsize=(25, 15))

plt.plot(df_Volkswagen.Engine_size, df_Volkswagen.Horsepower, 'o--b')

plt.xlabel('Engine Size')
plt.ylabel('Horsepower')
plt.title('Engine Size vs Horsepower')
plt.suptitle('VW')

plt.show()
```

```python
plt.figure(figsize=(25, 15))

plt.plot(df_Volkswagen.Engine_size, df_Volkswagen.Fuel_efficiency, 'o--b')

plt.xlabel('Engine Size')
plt.ylabel('Fuel Efficiency')
plt.title('Fuel Efficiency Per Engine Size')
plt.suptitle('VW')

plt.show()
```
