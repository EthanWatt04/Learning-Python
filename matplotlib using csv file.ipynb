```python
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
```

```python
gas = pd.read_csv('gas_prices.csv')

gas
```

```python
plt.figure(figsize=(10,5))

plt.plot(gas.Year, gas.USA, 'o-', label='USA')
plt.plot(gas.Year, gas.Canada, 'o-', label='Canada')
plt.plot(gas.Year, gas.Italy, 'o-', label='Italy')
plt.plot(gas.Year, gas.France, 'o-', label='France')

plt.xticks(gas.Year[::2])

plt.xlabel('Year')
plt.ylabel('Price per Gallon')

plt.title('Gas Prices Over The Years')

plt.legend()

plt.show()

#plt.savefig('Gas_Price_figure', dpi=300)
```

```python
plt.figure(figsize=(10,5))

for country in gas:
    if country != 'Year':
        plt.plot(gas.Year, gas[country])
```
